```python
# Importamos el dataset de los datos de los pokemon
pokemon = pd.read_csv('dataset_final.csv')
pokemon.head()
```

```python
'''
1. ¿Cuántos pokémon hay en el dataset?
'''
n_pkmn = pokemon["Name"].count()
print(f"Hay {n_pkmn} pokémon en el dataset")
```

```python
'''
2. ¿Cuántos pokémon hay de tipo Poison? (Type 1)
'''
n_poison = pokemon[pokemon["Type 1"] == "Poison"]
print(f"Hay {len(n_poison)} pokémon tipo 1 Veneno")
```

```python
'''
3. ¿Cuántos tipos diferentes de pokémon hay? (Type 1)
'''
n_type = pokemon["Type 1"].unique()
print(f"Hay {len(n_type)} tipos de pokémon diferentes:")
for element in n_type:
	print(f"* {element}")
```

```python
'''
2. ¿Cuántos pokémon hay de cada tipo? (Type 1)
'''
max_type = n_type[0]
for type in n_type:
	n_of_type = pokemon[pokemon["Type 1"] == type]
	if (len(n_of_type)) > len(pokemon[pokemon["Type 1"] == max_type]):
		max_type = type
	print(f"Hay {len(n_of_type)} pokémon de tipo '{type}'")
'''
2.1 ¿cual es el tipo de pokémon con más pokémon? (Type 1)
'''
len_max_type = len(pokemon[pokemon["Type 1"] == max_type])
print(f"El tipo (tipo 1) con más pokémon es '{max_type}' con {len_max_type} pokémons")
```

```python
'''
3. ¿Cuál es el pokémon más rápido?
'''
sort_speed = pokemon.sort_values("Speed", ascending=False)
fastest_name = sort_speed['Name'].values[0]
fastest_speed = sort_speed['Speed'].values[0]
print(f"El pokémon más rápido es {fastest_name} con {fastest_speed} puntos de Velocidad")
```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100?
'''
defense_speed = pokemon[defense_condition & fast_condition]
print(f"{len(defense_speed)} pokémon tienen una defensa y velocidad de más de 100")
```

```python
'''
9. Calcula el promedio de stats de los pokemon Legendary (HP, Attack, Defense, Sp. Atk, Sp. Def, Speed) y los no Legendary
'''
no_legendary_condition = pokemon["Legendary"] == False
no_legendary_pkmn = pokemon[no_legendary_condition]

dict_legendary = dict()
dict_no_legendary = dict()

means_legendaray = legendary_pkmn.describe().loc[["mean"]]
print("Pokémons legendarios:")
for i in means_legendaray:
	if (i != "#" and i != "Total" and i != "Generation"):
		dict_legendary[i] = round(means_legendaray[i][0], 2)
		print(f"    * La media de {i} es: {round(means_legendaray[i][0], 2)}")

means_no_legendaray = no_legendary_pkmn.describe().loc[["mean"]]
print("Pokémons no legendarios:")
for i in means_no_legendaray:
	if (i != "#" and i != "Total" and i != "Generation"):
		dict_no_legendary[i] = round(means_no_legendaray[i][0], 2)
		print(f"    * La media de {i} es: {round(means_no_legendaray[i][0], 2)}")
```

```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

#Propio: Función que devuelve un diccionario con los articulos comprado y el precio total de cada uno (precio * cantidad)
def	total_comprado(articulos, compra): 
	comprado = dict()
	for item in compra:
		if item not in comprado:
			comprado[item] = articulos[item]
		else:
			comprado[item] += articulos[item]
	return comprado

# A. Una función que calcule el precio total de la compra
def total_price(articulos, compra):
	comprado = total_comprado(articulos, compra)
	precio = reduce(lambda total, art: total + comprado[art], comprado, 0)
	return precio

# B. Una función que calcule el precio total de la compra con un descuento del 10%
def ten_percent(articulos, compra):
	return (total_price(articulos, compra) * 0.9)

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
def condicional_percent(articulos, compra):
	if (total_price(articulos, compra) > 100):
		return ten_percent(articulos, compra)
	return total_price(articulos, compra)

# D. Una función que calcule el precio total aplicando el IVA (21%)
def	iva(articulos, compra):
	comprado = total_comprado(articulos, compra)
	for item in comprado:
		comprado[item] *= 1.21
	precio = reduce(lambda total, art: total + comprado[art], comprado, 0)
	return precio

# E.lista los artículos cuyo precio es superior a 20€
def price_more_twenty(articulos):
	more_twenty = list(filter(lambda x: articulos[x] > 20, articulos))
	return more_twenty
```
